### Binary Features: Flag, Bool, True-False

```python
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/htcso/OneDrive/Masaüstü/pythonProject/datasets/titanic.csv")
df.head()
```

```python
df["NEW_CABIN_BOOL"] = df["Cabin"].notnull().astype("int")
```

```python
df.head()
```

```python
df.groupby("NEW_CABIN_BOOL")["Survived"].mean()
```

```python
# Yeni oluşturulan değişken ile bağımlı değişken arasında anlamlı bir ilişki olup olmadığını anlamak için oran testi

from statsmodels.stats.proportion import proportions_ztest

test_stats, pvalue = proportions_ztest(count = [df.loc[df["NEW_CABIN_BOOL"] == 1, "Survived"].sum(),
                                                df.loc[df["NEW_CABIN_BOOL"] == 0, "Survived"].sum()],
                                        nobs = [df.loc[df["NEW_CABIN_BOOL"] == 1, "Survived"].shape[0],
                                                df.loc[df["NEW_CABIN_BOOL"] == 0, "Survived"].shape[0]])
print("Test Stats = %.4f, p-value = %.4f" % (test_stats, pvalue))
```

```python
# pvalue < 0.05 --> anlamlı bir fark vardır
```

```python
# Gemide yalnız olup olmamak hayatta kalma durumunu etkiliyor mu ?
```

```python
df.loc[(df["SibSp"] + df["Parch"] > 0), "NEW_IS_ALONE"] = "NO"
df.loc[(df["SibSp"] + df["Parch"] == 0), "NEW_IS_ALONE"] = "YES"
```

```python
df.head()
```

```python
df.groupby("NEW_IS_ALONE")["Survived"].mean()
```

```python
test_stats, pvalue = proportions_ztest(count = [df.loc[df["NEW_IS_ALONE"] == "YES", "Survived"].sum(),
                                                df.loc[df["NEW_IS_ALONE"] == "NO", "Survived"].sum()],
                                        nobs = [df.loc[df["NEW_IS_ALONE"] == "YES", "Survived"].shape[0],
                                                df.loc[df["NEW_IS_ALONE"] == "NO", "Survived"].shape[0]])
print("Test Stats = %.4f, p-value = %.4f" % (test_stats, pvalue))
```

```python
# pvalue < 0.05 --> anlamlı bir fark vardır
```

### Text Features

```python
# Letter Count
df["NEW_NAME_COUNT"] = df["Name"].str.len()
```

```python
df.loc[:, ["Name", "NEW_NAME_COUNT"]].head()
```

```python
# Word Count
df["NEW_NAME_WORD_COUNT"] = df["Name"].apply(lambda x: len(str(x).split(" ")))
```

```python
df.loc[:, ["Name", "NEW_NAME_WORD_COUNT"]].head()
```

```python
# Özel Yapıları Yakalamak
```

```python
# Doktor olanlar 1
df["NEW_NAME_DR"] = df["Name"].apply(lambda x: len([x for x in x.split() if x.startswith("Dr")]))
```

```python
df.loc[:, ["Name", "NEW_NAME_DR"]].head()
```

```python
df.groupby("NEW_NAME_DR")["Survived"].agg(["mean", "count"])
```

### Regex Features

```python
df["NEW_TITLE"] = df["Name"].str.extract(' ([A-Za-z]+)\.', expand=False)
```

```python
df.loc[:, ["Name", "NEW_TITLE"]].head()
```

```python
df.groupby("NEW_TITLE").agg({"Survived": "mean", "Age": ["count", "mean"]})
```

```python
dff["Timestamp"] = pd.to_datetime(dff["Timestamp"], format="%Y-%m-%d")
```

```python
# year
dff["Year"] = dff["Timestamp"].dt.year
```

```python
# month
dff["Month"] = dff["Timestamp"].dt.month
```

```python
# month diff
dff["month_diff"] = (date.today().year - dff["Timestamp"].dt.year) * 12 + date.today().month - dff["Timestamp"].dt.month
```

```python
# day name
dff["day_name"] = dff["Timestamp"].dt.day_name()
```

```python
df = pd.read_csv("C:/Users/htcso/OneDrive/Masaüstü/pythonProject/datasets/titanic.csv")
df.head()
```
